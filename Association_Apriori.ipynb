## Data Preprocessing

因Apriori演算法僅支援離散型資料進行分析，因此我們同樣需要將資料從連續型數值變成離散型資料

```python
import pandas as pd

# Continuous to Discrete
k=3

# 將資料分為k個區間
f1 = pd.cut(X['fixed acidity'], k)
f2 = pd.cut(X['volatile acidity'], k)
f3 = pd.cut(X['citric acid'], k)
f4 = pd.cut(X['residual sugar'], k)
f5 = pd.cut(X['chlorides'], k)
f6 = pd.cut(X['free sulfur dioxide'], k)
f7 = pd.cut(X['total sulfur dioxide'], k)
f8 = pd.cut(X['density'], k)
f9 = pd.cut(X['pH'], k)
f10 = pd.cut(X['sulphates'], k)
f11 = pd.cut(X['alcohol'], k)
f12 = pd.cut(X['quality'], k)
# 建立新的 DataFrame
data = {
    'fixed acidity': f1,
    'volatile acidity': f2,
    'citric acid': f3,
    'residual sugar': f4,
    'chlorides': f5,
    'free sulfur dioxide': f6,
    'total sulfur dioxide': f7,
    'density': f8,
    'pH': f9,
    'sulphates': f10,
    'alcohol': f11,
    'quality': f12,
    'quality category': y['quality category']
}

# 生成新的 DataFrame
winedata_df = pd.DataFrame(data)
print(winedata_df)
```

```python
type(winedata_df['fixed acidity'][0])
```

```python
winedata_str = winedata_df.astype(str)
type(winedata_str['fixed acidity'][0])
```

```python
# 為每個欄位添加後綴
# 為每個欄位添加後綴
# 為每個欄位添加後綴
winedata_str['fixed acidity'] = list(map(lambda x: x + '_fa', winedata_str['fixed acidity']))
winedata_str['volatile acidity'] = list(map(lambda x: x + '_va', winedata_str['volatile acidity']))
winedata_str['citric acid'] = list(map(lambda x: x + '_ca', winedata_str['citric acid']))
winedata_str['residual sugar'] = list(map(lambda x: x + '_rs', winedata_str['residual sugar']))
winedata_str['chlorides'] = list(map(lambda x: x + '_cl', winedata_str['chlorides']))
winedata_str['free sulfur dioxide'] = list(map(lambda x: x + '_fsd', winedata_str['free sulfur dioxide']))
winedata_str['total sulfur dioxide'] = list(map(lambda x: x + '_tsd', winedata_str['total sulfur dioxide']))
winedata_str['density'] = list(map(lambda x: x + '_d', winedata_str['density']))
winedata_str['pH'] = list(map(lambda x: x + '_ph', winedata_str['pH']))
winedata_str['sulphates'] = list(map(lambda x: x + '_su', winedata_str['sulphates']))
winedata_str['alcohol'] = list(map(lambda x: x + '_alc', winedata_str['alcohol']))
winedata_str['quality'] = list(map(lambda x: x + '_qual', winedata_str['quality']))


winedata_str
```

```python
winedata_str_arr = winedata_str.values
type(winedata_str_arr)
```

```python
winedata_str_arr
```

## Model Training

資料準備完成後，需要import Apriori演算法套件進行運算

```python
from apyori import apriori
```

```python
association_rules = apriori(winedata_str_arr, min_support=0.05, min_confidence=0.6, min_lift=1.2, max_length=4)
association_results = list(association_rules)
association_results
```
