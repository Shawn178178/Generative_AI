#1.引入模組以及準備資料

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
```

```python
row_data=pd.read_csv('winequality-red.csv')
row_data=pd.DataFrame(row_data)
X = row_data.iloc[:, :11+1]
y= row_data[['quality category']]

X=X.to_numpy()
```

#2.scikit learn

##A.引入模組

```python
from sklearn.cluster import AgglomerativeClustering
```

##B.進行分群

```python
ml=AgglomerativeClustering(n_clusters=3,metric='euclidean',linkage='ward')
```

```python
ml.fit_predict(X)
```

##C.分群結果與圖像化顯示

```python
plt.scatter(X[:,7],X[:,10],c=ml.fit_predict(X))
```

#3.Scipy

##A.引入層次聚集的模型

```python
import scipy.cluster.hierarchy as sch
```

##B.進行分群，並圖像化整個樹狀圖

```python
dis=sch.linkage(X,metric='euclidean',method='ward')
```

```python
sch.dendrogram(dis)
plt.title('Hierarchical Clustering')
plt.show()
x_axis_labels.txt
```

##C.利用距離決定群數，或直接給定群數

###一.距離切割

```python
max_dis=5
clusters=sch.fcluster(dis,max_dis,criterion='distance')
clusters
```

```python
plt.scatter(X[:,10],X[:,11],c=clusters)
```

###二.直接給定群數

```python
k=5
clusters=sch.fcluster(dis,k,criterion='maxclust')
```

```python
plt.scatter(X[:,0],X[:,1],c=clusters)
```
